#Xóa stopword

```python
stopword_path = path+'/vietnamese_stopwords.txt'
with open(stopword_path, 'r', encoding="utf8") as file:
    stopwords = [word.strip() for word in file.readlines()]

def remove_stopword(data, stopwords):
    list_re = []
    for item in data:
        words = item.split()  # Tách từ trong văn bản thành danh sách các từ
        filtered_words = [word for word in words if word.lower() not in stopwords]  # Loại bỏ stopword
        filtered_text = " ".join(filtered_words)  # Chuyển danh sách các từ đã lọc thành văn bản mới
        list_re.append(filtered_text)
    return list_re

list1 = []
list2 = []
list1 = remove_stopword(list_title_1d, stopwords)
list2 = remove_stopword(list_des_1d, stopwords)
```

```python
X = pd.DataFrame({'title': list1, 'Des': list2})
```
