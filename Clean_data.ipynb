```python
def Lam_sach_du_lieu(df):
  columns = ['title', 'Des']
  words =['(Dân trí) - ', '\n', '"', '\t', ':', '’', "'"]
  n = df.shape[0]
  for i in range(n):
    for column in columns:
      for word in words:
        if word in df.loc[i, column]:
          df.loc[i, column] = df.loc[i, column].replace(word, "")
  words =['. ', '?', '…', '!']
  for i in range(n):
    for column in columns:
      for word in words:
        if word in df.loc[i, column]:
          df.loc[i, column] = df.loc[i, column].replace(word, ", ")
  return df
```

```python
df = df.drop_duplicates(subset=['title']).reset_index(drop=True)
df2 = df2.drop_duplicates(subset=['title']).reset_index(drop=True)
df3 = df3.drop_duplicates(subset=['title']).reset_index(drop=True)
df = df.drop_duplicates(subset=['Des']).reset_index(drop=True)
df2 = df2.drop_duplicates(subset=['Des']).reset_index(drop=True)
df3 = df3.drop_duplicates(subset=['Des']).reset_index(drop=True)
```

```python
df = Lam_sach_du_lieu(df)
df2 = Lam_sach_du_lieu(df2)
df3 = Lam_sach_du_lieu(df3)
```
