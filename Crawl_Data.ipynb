```python
# i = 1
# đổi khoa học thành khoa học và công nghệ
for i in range(1,5):
    TopicLists = [
        f'https://vnexpress.net/thoi-su-p{i}',
        f'https://vnexpress.net/kinh-doanh-p{i}',
        f'https://vnexpress.net/bat-dong-san-p{i}',
        f'https://vnexpress.net/khoa-hoc-p{i}',
        f'https://vnexpress.net/so-hoa-p{i}',
        f'https://vnexpress.net/giai-tri-p{i}',
        f'https://vnexpress.net/the-thao-p{i}',
        f'https://vnexpress.net/giao-duc-p{i}',
        f'https://vnexpress.net/suc-khoe-p{i}',
    ]
    for j in range(0,9):
        url = TopicLists[j]
        # print(url)
        soup = get_page_content(url)
        Title =[]
        Link = []
        Des = []
        Topic =[]
        all_h3 = soup.find_all('h3',{'class':'title-news'})
        all_h2 = soup.find_all('h2',{'class':'title-news'})
        for news_i in range(len(all_h3)):
            topic_tmp = ""
            des_tmp = ""
            a = all_h3[news_i].find('a')
            Title.append(a.get_text())
            # print(title)
            link_tmp = a.get('href')
            # Link.append(link_tmp)
            news_link = f'{link_tmp}'
            news_soup_content = get_page_content(news_link)
            news_all_des = news_soup_content.find_all('p',{'class':'description'})
            news_all_topic = news_soup_content.find_all('ul',{'class':'breadcrumb'})
            for des_i in range(len(news_all_des)):
                des_tmp = news_all_des[des_i].get_text(separator=' ')
            for topic_i in range(len(news_all_topic)):
                news_topic = news_all_topic[0].find('a')
                topic_tmp = news_topic.get_text()
            Topic.append(topic_tmp)
            Des.append(des_tmp)

        for news_i_1 in range(len(all_h2)):
            topic_tmp = ""
            des_tmp = ""
            a = all_h2[news_i_1].find('a')
            Title.append(a.get_text())
            # print(title)
            link_tmp = a.get('href')
            # Link.append(link_tmp)
            news_link = f'{link_tmp}'
            news_soup_content = get_page_content(news_link)
            news_all_des = news_soup_content.find_all('p',{'class':'description'})
            news_all_topic = news_soup_content.find_all('ul',{'class':'breadcrumb'})
            for des_i in range(len(news_all_des)):
                des_tmp = news_all_des[des_i].get_text(separator=' ')
            for topic_i in range(len(news_all_topic)):
                news_topic = news_all_topic[0].find('a')
                topic_tmp = news_topic.get_text()
            Topic.append(topic_tmp)
            Des.append(des_tmp)
        df_tmp = pd.DataFrame({'title':Title,'Des':Des,'Topic':Topic})
        df_Vnexp = pd.concat([df_Vnexp, df_tmp], axis = 0, ignore_index=True)


```
